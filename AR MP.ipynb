### Numerical verification of AR-MP distribution

```julia
model = ARMPModel(0.2, 0.7)
l, u = lb(model), ub(model)
println(l);
println(u);
```

```julia
sampleSpec = randSpec(model, 10)

bins = linspace(minimum(sampleSpec), maximum(sampleSpec), 101)
_, counts = hist(sampleSpec, bins)
counts /= sum(counts) * mean(diff(bins))

x = logspace(log10(l + 1e-6), log10(u - 1e-6), 100)
l1 = layer(x=(bins[1:end - 1] + bins[2:end]) / 2, y=counts, Geom.bar)
lineTheme = Theme(line_width=1mm, default_color=color("red"))
l2 = layer(x=x, y=map(spec(model), x), Geom.line, lineTheme)
plot(l1, l2)
```

### Limiting behavior of AR-MP

```julia
phi = 1.0 - 1.0 / 2.0
sigma = 1 / sqrt(1 - phi^2)
p = 100; n = 10000;
```

```julia
armp_model = ARMPModel(p, n, phi)
mp_model = MPModel(p, n, sigma)

armp_l, armp_u = lb(armp_model), ub(armp_model)
armp_x = logspace(log10(armp_l + 1e-6), log10(armp_u - 1e-6), 101)
armp_plot = layer(x=armp_x, y=map(spec(armp_model), armp_x), Geom.line, Theme(default_color=color("red")))

mp_l, mp_u = lb(mp_model), ub(mp_model)
mp_x = logspace(log10(mp_l + 1e-6), log10(mp_u - 1e-6), 101)
mp_plot = layer(x=mp_x, y=map(spec(mp_model), mp_x), Geom.line)

plot(armp_plot, mp_plot)
```
