```julia
using Gadfly

include("armp.jl")
using ARMP
```

### Solve for the lower and upper eigenvalue bounds of the correlated noise

```julia
model = ARMPModel(0.5, 0.2)
l, u = lb(model), ub(model)
println(l);
println(u);
```

```julia
x = rand(model)
spy(x)
```

```julia
sampleSpec = randSpec(model, 10)

x = linspace(l + 1e-6, u - 1e-6, 20)
plot(x=sampleSpec, Geom.histogram)
```

### Numerical verification

```julia
phi = 0.1;
cs = linspace(0.01, 0.99, 101);
cData, dData = Float64[], Float64[];
for c in cs[1:10:end]
    N = 200; T = int(N / c);
    cData = [cData, c * ones(N)];
    dData = [dData, sampleSV2(N, T, phi, 1)];
end
plot(
layer(x=cs, y=map(c -> lb(c, phi), cs), Theme(default_color=color("blue")), Geom.line),
layer(x=cs, y=map(c -> ub(c, phi), cs), Theme(default_color=color("red")), Geom.line),
layer(x=cData, y=dData, Theme(default_point_size=0.5mm, highlight_width=0mm), Geom.point),
Guide.XLabel("N / T"), Guide.YLabel("Eigval"), Scale.y_continuous)
```
