```julia
using Gadfly
using Optim
```

```julia
function sampleAR(N, T, alpha)
    p = sqrt(1 - alpha^2) * eye(N);
    X = zeros(N, T)
    X[:, 1] = p * randn(N);
    for t in 2:T
        X[:, t] = alpha * X[:, t - 1] + randn(N);
    end
    return X;
end
```

```julia
function sampleSV2(N, T, alpha, nTrial)
    D = Float64[];
    for kTrial in 1:nTrial
        X = sampleAR(N, T, alpha);
        d, _ = eig(X * X' / T);
        D = [D, d];
    end
    return D;
end
```

```julia
function zFunc(c, phi)
    return s -> (-1 ./ s + 1 ./ sqrt((c * s + 1 + phi^2).^2 - 4 * phi^2));
end;

function ub(c, phi)
    f = zFunc(c, phi);
    # upper bound
    bounds = [-1 / c * (1 - phi)^2 + 1e-9, 0 - 1e-9];
    return optimize(f, bounds[1], bounds[2]).f_minimum;
end;

function lb(c, phi)   
    f = zFunc(c, phi);
    if c >= 1
        bounds = [-eps(Float64), -1 / c * (1 + phi)^2 - eps(Float64)];
    else
        bounds = [eps(Float64), 1e16];
    end
    return -optimize(x -> 0.0 - f(x), bounds[1], bounds[2]).f_minimum;
end;
```

```julia
phi = 0.3;
cs = linspace(0.05, 0.95, 101);
cData, dData = Float64[], Float64[];
for c in cs[1:10:end]
    N = 200; T = int(N / c);
    cData = [cData, c * ones(N)];
    dData = [dData, sampleSV2(N, T, phi, 1)];
end
plot(
layer(x=cs, y=map(c -> lb(c, phi), cs), Theme(default_color=color("blue")), Geom.line),
layer(x=cs, y=map(c -> ub(c, phi), cs), Theme(default_color=color("red")), Geom.line),
layer(x=cData, y=dData, Theme(default_point_size=0.5mm, highlight_width=0mm), Geom.point),
Guide.XLabel("N / T"), Guide.YLabel("Eigval"))
```
