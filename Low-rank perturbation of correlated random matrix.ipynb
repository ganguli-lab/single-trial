### Uncorrelated noise

```julia
M = 500; P = 2000; K = 40;
nTrial = 2;
sY = Float64[]; sX = Float64[];
for kTria in 1:nTrial
    U, _ = qr(randn(M, M));
    V, _ = qr(randn(P, P));
    Y = U[:, 1:K] * diagm((rand(K)) * 4) * V[:, 1:K]';
    X = Y + randn(M, P) / sqrt(1000);
    
    _, sy, _ = svd(Y); _, sx, _ = svd(X)
    sY = [sY, sort(sy)]; sX = [sX, sort(sx)];
end
```

```julia
plot(x=sY, y=sX, Geom.point, 
Theme(default_color=color("red"), default_point_size=1mm, panel_fill=color("black")), 
Guide.xlabel("Noiseless Singular Values"), Guide.ylabel("Noisy Singular Values"))
```

### How is the recovered Kth eigenvector different in the correlated slow mode and uncorrellated slow mode case?

```julia
function trial(p, n, k, fast_phi, slow_phi; uc=false)
    # embedding
    u, _ = qr(randn((p, p)))
    
    # slow and fast data
    fastModel = ARMPModel(p - k, n, fast_phi)
    slowModel = ARMPModel(k, n, slow_phi)
    fastX, slowX = rand(fastModel), rand(slowModel)
    
    X = u[:, 1:k] * slowX + u[:, k + 1:end] * fastX
    Xp = u[:, 1:k] * slowX[:, randperm(n)] + u[:, k + 1:end] * fastX
    return u, X, Xp
end
```

```julia
p, n = 500, 2000;
fast_phi = 0.1
slow_phis = linspace(fast_phi, 0.95, 25)
ks = 1:2:51

kth = zeros(length(slow_phis), length(ks))
k1th = zeros(length(slow_phis), length(ks))
angle = zeros(length(slow_phis), length(ks))

kthuc = zeros(length(slow_phis), length(ks))
k1thuc = zeros(length(slow_phis), length(ks))
angleuc = zeros(length(slow_phis), length(ks))
for (kslow, slow_phi) in enumerate(slow_phis)
    for (kk, k) in enumerate(ks)
        U, X, Xp = trial(p + k, n, k, fast_phi, slow_phi)
        
        e, v = eigs(X * X' / n; nev=k + 1)
        kth[kslow, kk], k1th[kslow, kk] = e[k], e[k + 1]
        angle[kslow, kk] = ccdf(Chisq(k), norm(v[:, k]' * U[:, 1:k] * U[:, 1:k]')^2 * (p + k) / k)
        
        e, v = eigs(Xp * Xp' / n; nev=k + 1)
        kthuc[kslow, kk], k1thuc[kslow, kk] = e[k], e[k + 1]
        angleuc[kslow, kk] = ccdf(Chisq(k), norm(v[:, k]' * U[:, 1:k] * U[:, 1:k]')^2 * (p + k) / k)
    end
end
```

```julia
spy(angle)
```

```julia
spy(angleuc)
```

```julia
spy((angleuc - angle) .> 0.0)
```

### How do things depend on the number of neurons and length of recordings?

```julia
p, k = 500, 5
ns = linspace(210, 1500, 40)
ms = linspace(10, 500, 40)

fast_phi = 0.1
slow_phi = 0.9

anglem = zeros(length(ms), length(ns))
angleucm = zeros(length(ms), length(ns))
for (km, m) in enumerate(ms)
    for (kn, n) in enumerate(ns)
        U, X, Xp = trial(p + k, int(n), k, fast_phi, slow_phi)
        ix = randperm(p + k)[1:int(m)]
        Um, Xm, Xpm = U[ix, :], X[ix, :], Xp[ix, :]
        Um, _ = qr(Um)
        
        e, v = eigs(Xm * Xm' / n; nev=k + 1)
        anglem[km, kn] = norm(v[:, k]' * Um[:, 1:k] * Um[:, 1:k]')
        
        e, v = eigs(Xpm * Xpm' / (n + k); nev=k + 1)
        angleucm[km, kn] = norm(v[:, k]' * Um[:, 1:k] * Um[:, 1:k]')
    end
end
```

```julia
spy(anglem)
```

```julia
spy(angleucm)
```

```julia
spy((angleucm .- anglem) ./ angleucm)
```
