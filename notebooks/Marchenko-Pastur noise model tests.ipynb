## 1. Testing functions that compute the spectra of Marchenko-Pastur noise model

Note that only the non-zero part of the spectrum is considered regardless of wide/narrow matrices

```julia
ns = [1000, 2000, 5000]
ps = 100:100:5000

f = figure(figsize=(8, 2))

@manipulate for n in ns, p in ps, σ in 0.1:0.1:1, bins in [20, 50, 100, 200]
    m = MPModel(p, n, σ)
    
    Ss = sv_rand(m)
    Su, Sl = sv_ub(m), sv_lb(m)
    
    Es = ev_rand(m)
    Eu, El = ev_ub(m), ev_lb(m)
    
    svspec = sv_spec(m)
    evspec = ev_spec(m)
    
    withfig(f) do
        x = ev_linsupport(m, 101)
        y = map(evspec, x)
        subplot(121)
        plt.hist(Es, bins, normed=true, histtype="step")
        plt.plot(x, y, "r--", linewidth="2")
        axvline(El, linewidth=2, color="k", linestyle="--")
        axvline(Eu, linewidth=2, color="k", linestyle="--")
        xticks([El, Eu]); yticks(); title("Eigenvalue Spectrum")
        
        x = sv_linsupport(m, 101)
        subplot(122)
        plt.hist(Ss, bins, normed=true, histtype="step")
        plt.plot(x, map(svspec, x), "r--", linewidth="2")
        axvline(Sl, linewidth=2, color="k", linestyle="--")
        axvline(Su, linewidth=2, color="k", linestyle="--")
        xticks([Sl, Su]); yticks(); title("Singular Value Spectrum")
    end
end
```

## 2. Testing functions that deals with the low-rank perturbations to a Marchenko-Pastur noise model

### a. Singular and eigen value transfer function checks

```julia
g = figure(figsize=(8, 3))

npt = 100

@manipulate for n in [500, 1000, 2000], p = 100:100:500, k = 1:5, σ in 0.1:0.1:1
    m = MPModel(p, n, σ)
    
    sthresh = sv_sigthresh(m)
    sv_out = zeros(npt, k)
    ev_out = zeros(npt, k)
    sv_in = linspace(0, sthresh * 2, npt)
    for (ksv, sv) in enumerate(sv_in)
        X = rand(m) + sv * qr(randn(p, k))[1] * qr(randn(n, k))[1]'
        ev_out[ksv, :] = eigs(X * X'; nev=k)[1]
        sv_out[ksv, :] = sqrt(ev_out[ksv, :])
    end
    
    sv_out_theory = map(x -> sv_xfer(m, x), sv_in)
    ev_out_theory = map(x -> ev_xfer(m, x), sv_in.^2)
    
    withfig(g) do
        subplot(121)
        plot(sv_in, sv_out, ".")
        plot(sv_in, sv_out_theory, "k-", linewidth=2)
        axvline(sthresh, color="r", linestyle="--", linewidth=2)
        axhline(sv_xfer(m, sthresh), color="r", linestyle="--", linewidth=2)
        subplot(122)
        plot(sv_in.^2, ev_out, ".")
        plot(sv_in.^2, ev_out_theory, "k-", linewidth=2)
        axvline(sthresh^2, color="r", linestyle="--", linewidth=2)
        axhline(sv_xfer(m, sthresh)^2, color="r", linestyle="--", linewidth=2)
    end
end
```
