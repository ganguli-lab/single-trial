```julia
push!(LOAD_PATH, "../src/")

using HDStat, PyPlot, Interact
```

```julia
n = 1000
ps = 100:100:5000
println(string("n: ", n))

f = figure(figsize=(8, 3))

@manipulate for p in ps
    m = MPModel(p, n, 1)
    X = rand(m)
    
    Ss = svd(X)[2]
    Su, Sl = sv_ub(m), sv_lb(m)
    
    if p < n
        Es = eig(X * X')[1]
    else
        Es = eig(X' * X)[1]
    end
    Eu, El = ev_ub(m), ev_lb(m)
    
    svspec = sv_spec(m)
    evspec = ev_spec(m)
    
    withfig(f) do
        x = ev_linsupport(m, 101)
        y = map(evspec, x)
        subplot(121)
        plt.hist(Es, 50, normed=true, histtype="step")
        plt.plot(x, y, "r-")
        axvline(El, linewidth=2, color="k", linestyle="--")
        axvline(Eu, linewidth=2, color="k", linestyle="--")
        xticks([El, Eu]); yticks(); title("Eigenvalue Spectrum")
        
        x = sv_linsupport(m, 101)
        subplot(122)
        plt.hist(Ss, 50, normed=true, histtype="step")
        plt.plot(x, map(svspec, x), "r-")
        axvline(Sl, linewidth=2, color="k", linestyle="--")
        axvline(Su, linewidth=2, color="k", linestyle="--")
        xticks([Sl, Su]); yticks(); title("Singular Value Spectrum")
    end
end
```
