```julia
include("HDStat.jl")

using PyPlot, HDStat, Distributions, Interact

# canonical angle of V into U
function cca(U, V)
    return sum(svd(U' * V)[2]) / size(V, 2)
end;
```

### Uncorrelated noise perturbed by rank-1 signal

```julia
figure(figsize=(6, 6))


@manipulate for p = 100:100:1000
    n = 1000; nTrial = 100;
    model = MPModel(p, n, 1)
    overlap = vecOverlap(model)
    xfer = eigvalXfer(model)

    ss = linspace(0.1, 2, nTrial)
    sp, lOverlap, rOverlap = zeros(nTrial), zeros(nTrial), zeros(nTrial)
    spTheory, lOverlapTheory, rOverlapTheory = zeros(nTrial), zeros(nTrial), zeros(nTrial)
    for (ks, s) in enumerate(ss)
        Z = rand(model)
        u, _ = qr(randn(p, 1))
        v, _ = qr(randn(n, 1))

        Y = u * s * sqrt(n) * v' + Z
        # left
        d, a = eigs(Y * Y' / n; nev=1)
        # right
        _, b = eigs(Y' * Y / n; nev=1)

        sp[ks] = d[1]; lOverlap[ks] = abs(sum(a .* u)); rOverlap[ks] = abs(sum(b .* v))
        lOverlapTheory[ks], rOverlapTheory[ks] = overlap(s^2)
        spTheory[ks] = xfer(s^2)
    end

    subplot(311)
    plot(ss, rOverlap, "ro")
    plot(ss, rOverlapTheory, "k", linewidth=2)
    ylim([0, 1])
    ylabel("Right svec")
    subplot(312)
    plot(ss, lOverlap, "ro")
    plot(ss, lOverlapTheory, "k", linewidth=2)
    ylim([0, 1])
    ylabel("Left svec")
    subplot(313)
    plot(ss, sqrt(sp), "ro")
    plot(ss, sqrt(spTheory), "k", linewidth=2)
    xlabel("Signal singular value")
    ylabel("Output sval")
end
```

### rank-$K$ perturbation

  - The signal model has $K$ identical singular values
  - The subspace overlap between the $K$-dimensional data and signal subspaces is lower bounded by the singular vector overlap of a rank-1 perturbation of the same signal power
  - This is a lower bound, because the theory provides only overlap between corresponding pairs of singular vectors

```julia
figure(figsize=(6, 6))

@manipulate for p = 100:100:1000
    n = 2000; k = 10; nTrial = 100;
    model = MPModel(p, n, 1)
    overlap = vecOverlap(model)
    xfer = eigvalXfer(model)

    ss = linspace(0.1, 2, nTrial)
    sp, lOverlap, rOverlap = zeros(nTrial, k), zeros(nTrial), zeros(nTrial)
    spTheory, lOverlapTheory, rOverlapTheory = zeros(nTrial), zeros(nTrial), zeros(nTrial)
    for (ks, s) in enumerate(ss)
        Z = rand(model)
        u, _ = qr(randn(p, k))
        v, _ = qr(randn(n, k))

        Y = u * diagm(s * sqrt(n) * ones(k)) * v' + Z
        # left
        d, a = eigs(Y * Y' / n; nev=k)
        # right
        _, b = eigs(Y' * Y / n; nev=k)

        sp[ks, :] = d; lOverlap[ks] = cca(u, a); rOverlap[ks] = cca(v, b)
        lOverlapTheory[ks], rOverlapTheory[ks] = overlap(s^2)
        spTheory[ks] = xfer(s^2)
    end

    subplot(311)
    plot(sqrt(ss), rOverlap, "ro")
    plot(sqrt(ss), rOverlapTheory, "k", linewidth=2)
    ylim([0, 1])
    ylabel("Right svec subspace")
    subplot(312)
    plot(sqrt(ss), lOverlap, "ro")
    plot(sqrt(ss), lOverlapTheory, "k", linewidth=2)
    ylim([0, 1])
    ylabel("Left svec")
    subplot(313)
    for j in 1:k
        plot(sqrt(ss), sqrt(sp[:, j]), ".")
    end
    plot(sqrt(ss), sqrt(spTheory), "k-", linewidth=2)
    xlabel("Signal singular value")
    ylabel("Output sval")
end
```
